```python
from tahoe import execute_async
import pandas as pd
```

```python
dfs = pd.read_excel('top_searched_queries_stats.xlsx', sheet_name=None)
```

```python
df_top = dfs['1000 top searched normalized qu']
df_top.columns = df_top.iloc[0]
df_top = df_top.drop(df_top.index[0])
df_top['query'] = df_top['normalized_query']
```

```python
df_mid = dfs['1000 Torso searched normalized ']
df_mid.columns = df_mid.iloc[0]
df_mid = df_mid.drop(df_mid.index[0])
df_mid['query'] = df_mid['normalized_query']

df_tail = dfs['1000 tail searched normalized q']
df_tail.columns = df_tail.iloc[0]
df_tail = df_tail.drop(df_tail.index[0])
df_tail['query'] = df_tail['normalized_query']
```

```python
q = f"select query, norm, categories, category_names, weights from search.query_new_category_inference_20230307"
res = execute_async(q)
```

```python
df_v0 = pd.DataFrame(res, columns=['query', 'norm', 'categories', 'category_names', 'weights'])
```

```python
len(df_v0)
```

```python
df_top_merge = df_top.merge(df_v0.rename(columns={'norm': 'v0_norm', 'categories': 'v0_categories', 'category_names': 'v0_category_names', 
    'weights': 'v0_weights'
}), on='query', how='left')
```

```python
df_mid_merge = df_mid.merge(df_v0.rename(columns={'norm': 'v0_norm', 'categories': 'v0_categories', 'category_names': 'v0_category_names', 
    'weights': 'v0_weights'
}), on='query', how='left')

df_tail_merge = df_tail.merge(df_v0.rename(columns={'norm': 'v0_norm', 'categories': 'v0_categories', 'category_names': 'v0_category_names', 
    'weights': 'v0_weights'
}), on='query', how='left')
```

```python
q = f"select query, categories, category_names, weights from structured_data.query_top3_predictions where model_version = 2"
res = execute_async(q)
```

```python
df_v2 = pd.DataFrame(res, columns=['query', 'categories', 'category_names', 'weights'])
```

```python
len(df_v2)
```

```python
df_top_merge = df_top_merge.merge(df_v2.rename(columns={'categories': 'v2_categories', 'category_names': 'v2_category_names', 
    'weights': 'v2_weights'
}), on='query', how='left')
```

```python
df_mid_merge = df_mid_merge.merge(df_v2.rename(columns={'categories': 'v2_categories', 'category_names': 'v2_category_names', 
    'weights': 'v2_weights'
}), on='query', how='left')

df_tail_merge = df_tail_merge.merge(df_v2.rename(columns={'categories': 'v2_categories', 'category_names': 'v2_category_names', 
    'weights': 'v2_weights'
}), on='query', how='left')
```
