```python
import pandas as pd
```

```python
from sklearn.metrics import classification_report
```

```python
df = pd.read_json("../../data/wish_products/wish-mturk-labelled-09202022-clean-joinedlance-openai_reverse_fewshot_inferred.json", lines=True)
```

```python
df.head()
```

```python
print(classification_report(df.category.apply(lambda x: ' > '.join(x[:1])), 
    df.openai_reverse_predicted_categories.apply(lambda x: ' > '.join(x[0][0][:1])), zero_division=0))
```

```python
print(classification_report(df.category.apply(lambda x: ' > '.join(x[:1])), 
    df.lance_predicted_category.apply(lambda x: ' > '.join(x[:1])), zero_division=0))
```

```python
print(classification_report(df.category.apply(lambda x: ' > '.join(x[:2])), 
    df.openai_reverse_predicted_categories.apply(lambda x: ' > '.join(x[0][0][:2])), zero_division=0))
```

```python
print(classification_report(df.category.apply(lambda x: ' > '.join(x[:2])), 
    df.lance_predicted_category.apply(lambda x: ' > '.join(x[:2])), zero_division=0))
```
